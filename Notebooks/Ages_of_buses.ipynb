### Age of buses

Combining historic and modern datasets

All Data from https://www.gov.uk/government/statistical-data-sets/bus-statistics-data-tables#historical-data-downloads

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
plt.style.use("seaborn-colorblind")
plt.style.use("seaborn-poster")
```

```python
df = pd.read_excel("../Data/bus06.ods", sheet_name="BUS06f", header=6)
df.head()
```

```python
df[["Year purchased", "London", "England", "Scotland", "Wales", "England outside London"]].plot(x="Year purchased", kind="bar")
```

```python
df["_year"] = df["Year purchased"].str.extract('(\d+)').dropna().astype("int")
df.head()
```

```python
plt.plot(df._year, df.London*100, label="London")
plt.plot(df._year, df.Scotland*100, label="Scotland")
plt.plot(df._year, df.Wales*100, label="Wales")
plt.plot(df._year, df["England outside London"]*100, label="England outside London")
plt.title("Distribution of Bus Purchase Year")
plt.xlabel("Year Purchased")
plt.ylabel("Percentage of Buses")
plt.legend()
```
